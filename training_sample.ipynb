```python
import numpy as np
import scipy.io
from sklearn import preprocessing
from sklearn.model_selection import train_test_split   # to split the dataset
import ssvm
```

```python
# load the data together with labels and split it into training and testing

X = scipy.io.loadmat('data/Trondheim_2023.mat')['matrix']      # since mat file has dict = matrix
y = np.load('data/trondheim_2023-03-27_0939Z-labels.npy')

X = preprocessing.scale(X, axis=0)                    # Normalization
# X = preprocessing.normalize(X, axis=0)             # Standardization

y[y == 0] = 2
print(X.shape, y.shape)

# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.98, random_state=1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# run the SSVM classifier on training dataset with user defined parameters

ssvm1 = ssvm.SSVM(eps=0.1, beta=1.96, delta=1.96, verbose=True, Lambda = 0.2)       
sparse_SVM = ssvm1.fit(X_train, y_train, track = True)
```

```python
# save the classifier

ssvm1.save('Trondheim2023.json')
```

```python
# check training accuracy

x_train_guess = ssvm1.predict(X_train)
print('Total number of data for training =', ssvm1.NSVM)
correct_classification = x_train_guess - y_train
print('Correctly classified =',len(correct_classification[np.where(correct_classification == 0)]))
print('Classification (in %) =', len(correct_classification[np.where(correct_classification == 0)])*100/ssvm1.NSVM)
```

```python
# check testing accuracy

x_test_guess = ssvm1.predict(X_test)
correct_prediction = x_test_guess - y_test
Ntest = len(y_test)
print('Total number of data for testing =', Ntest)
print('correctly predicted =',len(correct_prediction[np.where(correct_prediction == 0)]))
print('Accuracy (in %) =', len(correct_prediction[np.where(correct_prediction == 0)])*100/Ntest)
```

```python
# check full data accuracy

x_full_guess = ssvm1.predict(X)
correct_prediction = x_full_guess - y
Ntest = len(y)
print('Total number of data for testing =', Ntest)
print('correctly predicted =',len(correct_prediction[np.where(correct_prediction == 0)]))
print('Accuracy (in %) =', len(correct_prediction[np.where(correct_prediction == 0)])*100/Ntest)
```

```python
# plot the sparsity of each classifier corresponding to the class 

import matplotlib.pyplot as plt
I = [2, 4, 5, 6, 9]
for i in range(5):
    plt.plot(ssvm1._w()[I[i]],label='Class %s' %I[i])
    
plt.legend(title='sparsity')
plt.xlabel("Number of bands")
plt.ylabel("Weights")
plt.savefig("SVM_weights_Trondheim.pdf")
```
